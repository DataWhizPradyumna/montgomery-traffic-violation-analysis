```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
from tqdm.notebook import tqdm_notebook  # If you're using Jupyter

# Initialize tqdm for pandas operations
tqdm.pandas()

# First, let's handle datetime columns properly
print("Processing datetime columns...")
df['Date Of Stop'] = pd.to_datetime(df['Date Of Stop'])
df['Time Of Stop'] = pd.to_datetime(df['Time Of Stop'], format='%H:%M:%S').dt.time

# Create features for time analysis (for research question 4)
print("Creating time-based features...")
df['Hour'] = pd.to_datetime(df['Time Of Stop'], format='%H:%M:%S').dt.hour
df['Day_of_Week'] = df['Date Of Stop'].dt.day_name()
df['Month'] = df['Date Of Stop'].dt.month
df['Year_of_Stop'] = df['Date Of Stop'].dt.year

# Strategy 1: Categorical columns with low missingness
categorical_low_missing = ['Color', 'Make', 'Model', 'State', 'Driver State', 'Description', 'Location']

print("Imputing categorical columns with low missingness...")
for col in tqdm(categorical_low_missing, desc="Imputing categorical columns"):
    most_common = df[col].mode()[0]
    df[col].fillna(most_common, inplace=True)

# Strategy 2: For DL State, use Driver State where missing
print("Imputing DL State...")
df['DL State'] = df.progress_apply(lambda x: x['Driver State'] if pd.isna(x['DL State']) else x['DL State'], axis=1)

# Strategy 3: For Driver City, use location information
print("Imputing Driver City...")
df['Driver City'].fillna('UNKNOWN', inplace=True)

# Strategy 4: Vehicle Year - use median by Make and Model where possible
print("Imputing Vehicle Year...")
make_model_groups = df.groupby(['Make', 'Model'])
df_groups = [group for _, group in tqdm(make_model_groups, desc="Processing Make-Model groups")]

for group in tqdm(df_groups, desc="Imputing Year by Make-Model"):
    median_year = group['Year'].median()
    if not np.isnan(median_year):
        idx = group.index[group['Year'].isna()]
        df.loc[idx, 'Year'] = median_year

# Fill remaining NAs with overall median
median_year = df['Year'].median()
df['Year'].fillna(median_year, inplace=True)

# Strategy 5: Article - related to legal charges, best to add a "MISSING" category
print("Imputing Article...")
df['Article'].fillna('MISSING', inplace=True)

# Strategy 6: For the search-related columns with extremely high missingness
print("Processing search-related columns...")
df['Search Conducted'] = df['Search Conducted'].fillna('N')

search_cols = ['Search Disposition', 'Search Outcome', 'Search Reason', 
              'Search Reason For Stop', 'Search Type', 'Search Arrest Reason']

# For these, fill with "NOT_APPLICABLE" when Search Conducted is N or NA
for col in tqdm(search_cols, desc="Imputing search-related columns"):
    df[col] = df.progress_apply(
        lambda x: "NOT_APPLICABLE" if (x['Search Conducted'] == 'N' or pd.isna(x['Search Conducted'])) 
                 else x[col], 
        axis=1
    )

# Convert relevant columns to appropriate types
print("Converting column types...")
# Numeric columns
numeric_cols = ['Year']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Encode categorical variables for modeling
print("Encoding categorical variables...")
categorical_cols = ['Race', 'Gender', 'Violation Type', 'SubAgency', 'VehicleType']
for col in tqdm(categorical_cols, desc="Encoding categorical variables"):
    # Create label encoders
    le = LabelEncoder()
    # Fit on non-null values
    non_null_values = df[col].dropna()
    le.fit(non_null_values)
    # Transform only non-null values
    df.loc[~df[col].isna(), col + '_encoded'] = le.transform(non_null_values)

print("Saving processed dataframe...")
df.to_csv('processed_traffic_data.csv', index=False)
print("Data processing complete!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder



# Convert 'Accident' column from Yes/No to 1/0
# Based on your sample, 'Accident' is 'No' instead of a boolean
if 'Accident' in df.columns and df['Accident'].dtype == object:
    df['Accident'] = df['Accident'].map({'Yes': 1, 'No': 0})

# Create a contingency table of violation types vs accidents
accident_by_violation = pd.crosstab(df['Violation Type'], df['Accident'], 
                                   normalize='index') * 100

# Sort by accident rate (descending)
if 1 in accident_by_violation.columns:
    # If columns are numeric (1/0)
    accident_rates = accident_by_violation[1].sort_values(ascending=False)
else:
    # If not, we'll use the last column (assuming that's the "Yes" or positive case)
    accident_rates = accident_by_violation.iloc[:, -1].sort_values(ascending=False)

# Visualize the top 10 violation types with highest accident rates
plt.figure(figsize=(12, 8))
accident_rates.head(10).plot(kind='bar', color='crimson')
plt.title('Violation Types with Highest Accident Rates')
plt.xlabel('Violation Type')
plt.ylabel('Accident Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('accident_rates_by_violation.png')
plt.show()

# Statistical test for significance
from scipy.stats import chi2_contingency
# Create a contingency table without normalization
contingency_table = pd.crosstab(df['Violation Type'], df['Accident'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Conclusion: Significant association" if p < 0.05 else "Conclusion: No significant association")

# Predictive model for accidents based on violation type and other factors
# We need to properly encode categorical variables

# 1. First, let's identify continuous variables (already numeric) and categorical variables
all_columns = df.columns.tolist()
# Based on your sample data, these columns are likely categorical and need encoding
categorical_columns = ['Violation Type', 'VehicleType', 'SubAgency', 'Race', 'Gender']
# Keep only the categorical columns that actually exist in the dataframe
categorical_columns = [col for col in categorical_columns if col in all_columns]

# 2. For each categorical column, create dummy/one-hot variables
# This handles the string conversion issue
encoded_features = pd.get_dummies(df[categorical_columns], drop_first=False)

# 3. Add numeric columns that don't need encoding (check they exist and are numeric)
numeric_columns = ['Hour', 'Belts', 'Commercial Vehicle', 'HAZMAT', 'Work Zone']
# Filter to only include columns that exist and are already numeric
valid_numeric_columns = []
for col in numeric_columns:
    if col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            valid_numeric_columns.append(col)
        else:
            # For columns that exist but aren't numeric, map them
            if col in ['Belts', 'Commercial Vehicle', 'HAZMAT', 'Work Zone']:
                df[col] = df[col].map({'Yes': 1, 'No': 0}).fillna(0)
                valid_numeric_columns.append(col)

# 4. Combine encoded categorical features with numeric features
if valid_numeric_columns:
    features = pd.concat([encoded_features, df[valid_numeric_columns]], axis=1)
else:
    features = encoded_features

# 5. Target variable - make sure it's numeric
target = df['Accident']

# 6. Split data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# 7. Train model
print(f"Feature columns: {features.columns.tolist()[:5]}... (total: {len(features.columns)})")
print(f"Feature types: {features.dtypes.value_counts()}")
print(f"Target values: {target.value_counts()}")

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 8. Evaluate model
y_pred = rf_model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 9. Feature importance
importances = pd.DataFrame({
    'feature': features.columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='importance', y='feature', data=importances.head(15))
plt.title('Top 15 Most Important Features for Predicting Accidents')
plt.tight_layout()
plt.savefig('accident_feature_importance.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Start with the cleaned dataset
df = df_cleaned.copy()

print("RQ2: Predicting the likelihood of an arrest based on traffic stop data")
print("-" * 80)

# Create a binary arrest indicator if needed
if 'Was_Arrested' not in df.columns and 'Arrest Type' in df.columns:
    df['Was_Arrested'] = df['Arrest Type'].notna().astype(int)

print(f"Arrests in dataset: {df['Was_Arrested'].sum()} ({df['Was_Arrested'].mean()*100:.2f}%)")

# Check if we have multiple classes for meaningful analysis
if df['Was_Arrested'].nunique() == 1:
    print("\nWARNING: All records have the same arrest status. Focusing on descriptive analysis.")

# --- PART 1: Key factors associated with arrests ---

# 1. Violation Type Analysis
if 'Violation Type' in df.columns:
    print("\nAnalysis: Violation types and arrests")
    violation_arrest = pd.crosstab(df['Violation Type'], df['Was_Arrested'], normalize='index') * 100
    
    # If all arrests, just show top violation types
    if df['Was_Arrested'].nunique() == 1:
        violation_counts = df['Violation Type'].value_counts()
        print("\nTop 5 violation types in arrests:")
        for violation, count in violation_counts.head(5).items():
            print(f"  {violation}: {count} ({count/len(df)*100:.2f}%)")
    else:
        # Show arrest rates by violation type if we have variation
        arrest_rates = violation_arrest.iloc[:, -1].sort_values(ascending=False)
        print("\nTop 5 violation types by arrest rate:")
        for violation, rate in arrest_rates.head(5).items():
            print(f"  {violation}: {rate:.2f}%")
    
    # Statistical test
    contingency = pd.crosstab(df['Violation Type'], df['Was_Arrested'])
    chi2, p, dof, expected = chi2_contingency(contingency)
    print(f"\nChi-square test (Violation Type vs Arrests):")
    print(f"  Chi-square: {chi2:.2f}, p-value: {p:.6f}")
    print(f"  {'Significant association' if p < 0.05 else 'No significant association'} at α=0.05")
    
    # Create a single visualization
    plt.figure(figsize=(12, 6))
    
    if df['Was_Arrested'].nunique() == 1:
        # Bar chart of top violations
        violation_counts.head(10).plot(kind='bar', color='navy')
        plt.title('Top 10 Violation Types in Arrests')
        plt.xlabel('Violation Type')
        plt.ylabel('Count')
    else:
        # Bar chart of arrest rates by violation
        arrest_rates.head(10).plot(kind='bar', color='navy')
        plt.title('Violation Types with Highest Arrest Rates')
        plt.xlabel('Violation Type')
        plt.ylabel('Arrest Rate (%)')
    
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('rq2_violations_analysis.png')
    plt.show()

# 2. Time of Day Analysis
if 'Hour' in df.columns:
    print("\nAnalysis: Time of day and arrests")
    
    # Create hourly analysis
    hourly_data = df.groupby('Hour')['Was_Arrested'].agg(['count', 'sum', 'mean']).reset_index()
    hourly_data['arrest_rate'] = hourly_data['mean'] * 100
    
    # Find peak arrest hours
    peak_hours = hourly_data.sort_values('arrest_rate', ascending=False).head(3)
    print("\nPeak arrest hours:")
    for _, row in peak_hours.iterrows():
        print(f"  Hour {int(row['Hour'])}: {row['arrest_rate']:.2f}% arrest rate ({int(row['sum'])} arrests)")
    
    # Statistical test
    hourly_contingency = pd.crosstab(df['Hour'], df['Was_Arrested'])
    chi2, p, dof, expected = chi2_contingency(hourly_contingency)
    print(f"\nChi-square test (Hour vs Arrests):")
    print(f"  Chi-square: {chi2:.2f}, p-value: {p:.6f}")
    print(f"  {'Significant association' if p < 0.05 else 'No significant association'} at α=0.05")

# 3. Demographic Analysis
demographic_vars = []
for var in ['Race', 'Gender']:
    if var in df.columns:
        demographic_vars.append(var)

if demographic_vars:
    print("\nAnalysis: Demographics and arrests")
    
    for var in demographic_vars:
        # Create contingency table
        demo_contingency = pd.crosstab(df[var], df['Was_Arrested'])
        
        # Calculate arrest rates
        demo_rates = pd.crosstab(df[var], df['Was_Arrested'], normalize='index') * 100
        
        if df['Was_Arrested'].nunique() > 1:
            arrest_col = 1 if 1 in demo_rates.columns else demo_rates.columns[-1]
            print(f"\n{var} arrest rates:")
            for category, rate in demo_rates[arrest_col].sort_values(ascending=False).items():
                print(f"  {category}: {rate:.2f}%")
        else:
            demo_counts = df[var].value_counts()
            print(f"\n{var} distribution in arrests:")
            for category, count in demo_counts.items():
                print(f"  {category}: {count} ({count/len(df)*100:.2f}%)")
        
        # Statistical test
        chi2, p, dof, expected = chi2_contingency(demo_contingency)
        print(f"\nChi-square test ({var} vs Arrests):")
        print(f"  Chi-square: {chi2:.2f}, p-value: {p:.6f}")
        print(f"  {'Significant association' if p < 0.05 else 'No significant association'} at α=0.05")

# 4. Search-related factors
if 'Was_Searched' in df.columns:
    print("\nAnalysis: Searches and arrests")
    
    search_arrest_counts = pd.crosstab(df['Was_Searched'], df['Was_Arrested'])
    search_arrest_rates = pd.crosstab(df['Was_Searched'], df['Was_Arrested'], normalize='index') * 100
    
    if df['Was_Arrested'].nunique() > 1 and df['Was_Searched'].nunique() > 1:
        arrest_col = 1 if 1 in search_arrest_rates.columns else search_arrest_rates.columns[-1]
        print("\nArrest rates by search status:")
        print(f"  Searched: {search_arrest_rates.loc[True, arrest_col]:.2f}%")
        print(f"  Not searched: {search_arrest_rates.loc[False, arrest_col]:.2f}%")
    else:
        searched_count = df['Was_Searched'].sum()
        print(f"\nSearches conducted: {searched_count} ({searched_count/len(df)*100:.2f}% of arrests)")
    
    # Statistical test if possible
    if df['Was_Searched'].nunique() > 1:
        chi2, p, dof, expected = chi2_contingency(search_arrest_counts)
        print(f"\nChi-square test (Search vs Arrests):")
        print(f"  Chi-square: {chi2:.2f}, p-value: {p:.6f}")
        print(f"  {'Significant association' if p < 0.05 else 'No significant association'} at α=0.05")

# --- PART 2: Combined analysis of key factors ---

# Only if we have multiple classes and multiple variables
if df['Was_Arrested'].nunique() > 1 and 'Violation Type' in df.columns and 'Hour' in df.columns:
    print("\nMultivariate Analysis: Combined effect of violation type and time")
    
    # Get top violations
    top_violations = df['Violation Type'].value_counts().head(3).index.tolist()
    
    # Create a new variable combining hour (grouped) and violation
    # Group hours into meaningful periods (morning, afternoon, evening, night)
    df['Time_Period'] = pd.cut(
        df['Hour'], 
        bins=[0, 6, 12, 18, 24], 
        labels=['Night (0-6)', 'Morning (6-12)', 'Afternoon (12-18)', 'Evening (18-24)']
    )
    
    # Filter for top violations
    filtered_df = df[df['Violation Type'].isin(top_violations)]
    
    # Create the multi-factor contingency table
    multi_table = pd.crosstab(
        [filtered_df['Violation Type'], filtered_df['Time_Period']], 
        filtered_df['Was_Arrested'], 
        normalize='index'
    ) * 100
    
    # Print highest risk combinations
    if 1 in multi_table.columns:
        arrest_col = 1
    else:
        arrest_col = multi_table.columns[-1]
    
    print("\nHighest arrest rate combinations (violation type × time period):")
    sorted_rates = multi_table[arrest_col].sort_values(ascending=False).head(5)
    for (violation, time_period), rate in sorted_rates.items():
        print(f"  {violation} during {time_period}: {rate:.2f}% arrest rate")
    
    # Combined visualization
    plt.figure(figsize=(12, 8))
    
    # Create a heatmap of arrest rates
    pivot_data = pd.pivot_table(
        filtered_df, 
        values='Was_Arrested',
        index='Violation Type',
        columns='Time_Period',
        aggfunc='mean'
    ) * 100
    
    sns.heatmap(pivot_data, annot=True, fmt='.1f', cmap='YlOrRd', cbar_kws={'label': 'Arrest Rate (%)'})
    plt.title('Arrest Rates by Violation Type and Time Period')
    plt.tight_layout()
    plt.savefig('rq2_multivariate_analysis.png')
    plt.show()

print("\nResearch Question 2 Analysis Complete")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import statsmodels.api as sm
from statsmodels.formula.api import logit
import matplotlib.ticker as mtick

# Set plot style for better aesthetics
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Start with the cleaned dataset
df = df_cleaned.copy()

print("RQ3: How do traffic violations vary across different demographic groups?")
print("-" * 80)

# Check for required columns
if not all(col in df.columns for col in ['Race', 'Gender', 'Violation Type']):
    missing_cols = [col for col in ['Race', 'Gender', 'Violation Type'] if col not in df.columns]
    print(f"Error: Missing required columns: {missing_cols}")
else:
    # Display basic dataset information
    print(f"Total records: {len(df)}")
    print(f"Unique violation types: {df['Violation Type'].nunique()}")
    print(f"Unique racial groups: {df['Race'].nunique()}")
    print(f"Unique gender groups: {df['Gender'].nunique()}")
    
    # --- PART 1: Race and Violation Types ---
    print("\nPART 1: Analysis of Race and Violation Types")
    
    # Create contingency table
    race_violation = pd.crosstab(df['Race'], df['Violation Type'])
    
    # Chi-square test
    chi2, p, dof, expected = chi2_contingency(race_violation)
    print(f"\nChi-square test for Race vs. Violation Type:")
    print(f"Chi-square statistic: {chi2:.2f}")
    print(f"p-value: {p:.8f}")
    print(f"Degrees of freedom: {dof}")
    print(f"Conclusion: {'Significant association' if p < 0.05 else 'No significant association'} between race and violation types")
    
    # Get top 5 most common violation types for visualization
    top_violations = df['Violation Type'].value_counts().head(5).index.tolist()
    
    # Calculate percentages by race for each violation type
    race_violation_pct = pd.DataFrame()
    for violation in top_violations:
        if violation in race_violation.columns:
            # Calculate percentage within each race
            race_violation_pct[violation] = race_violation[violation] / race_violation.sum(axis=1) * 100
    
    # Display key differences
    print("\nViolation distribution by race (percentage within each racial group):")
    print(race_violation_pct.round(2))
    
    # Enhanced visualization: Top violation types by race
    plt.figure(figsize=(12, 8))
    
    # Prepare data in long format for better visualization
    race_violation_long = race_violation_pct.reset_index().melt(
        id_vars='Race', 
        value_vars=top_violations,
        var_name='Violation Type', 
        value_name='Percentage'
    )
    
    # Plot with seaborn for better aesthetics
    ax = sns.barplot(
        x='Race', 
        y='Percentage', 
        hue='Violation Type', 
        data=race_violation_long,
        palette='viridis'
    )
    
    # Enhance plot appearance
    plt.title('Distribution of Top 5 Violation Types by Race', fontweight='bold', pad=20)
    plt.xlabel('Race', labelpad=10)
    plt.ylabel('Percentage within Racial Group (%)', labelpad=10)
    plt.legend(title='Violation Type', title_fontsize=12, fontsize=10, loc='upper right', bbox_to_anchor=(1, 1))
    
    # Add percentage formatting to y-axis
    ax.yaxis.set_major_formatter(mtick.PercentFormatter())
    
    # Add a subtle grid for better readability
    ax.grid(axis='y', alpha=0.3)
    plt.xticks(rotation=45, ha='right')
    
    # Tight layout to ensure everything fits
    plt.tight_layout()
    plt.savefig('rq3_violations_by_race.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # --- PART 2: Gender and Violation Types ---
    print("\nPART 2: Analysis of Gender and Violation Types")
    
    # Create contingency table
    gender_violation = pd.crosstab(df['Gender'], df['Violation Type'])
    
    # Chi-square test
    chi2, p, dof, expected = chi2_contingency(gender_violation)
    print(f"\nChi-square test for Gender vs. Violation Type:")
    print(f"Chi-square statistic: {chi2:.2f}")
    print(f"p-value: {p:.8f}")
    print(f"Degrees of freedom: {dof}")
    print(f"Conclusion: {'Significant association' if p < 0.05 else 'No significant association'} between gender and violation types")
    
    # Calculate percentages by gender for each violation type
    gender_violation_pct = pd.DataFrame()
    for violation in top_violations:
        if violation in gender_violation.columns:
            # Calculate percentage within each gender
            gender_violation_pct[violation] = gender_violation[violation] / gender_violation.sum(axis=1) * 100
    
    # Display key differences
    print("\nViolation distribution by gender (percentage within each gender group):")
    print(gender_violation_pct.round(2))
    
    # Enhanced visualization: Top violation types by gender
    plt.figure(figsize=(10, 8))
    
    # Convert to long format
    gender_violation_long = gender_violation_pct.reset_index().melt(
        id_vars='Gender', 
        value_vars=top_violations,
        var_name='Violation Type', 
        value_name='Percentage'
    )
    
    # Create a more appealing plot
    ax = sns.barplot(
        x='Violation Type',
        y='Percentage',
        hue='Gender',
        data=gender_violation_long,
        palette='muted'
    )
    
    # Enhance plot appearance
    plt.title('Violation Types by Gender', fontweight='bold', pad=20)
    plt.xlabel('Violation Type', labelpad=10)
    plt.ylabel('Percentage within Gender Group (%)', labelpad=10)
    
    # Add percentage formatting to y-axis
    ax.yaxis.set_major_formatter(mtick.PercentFormatter())
    
    # Add a subtle grid for better readability
    ax.grid(axis='y', alpha=0.3)
    plt.xticks(rotation=30, ha='right')
    
    # Add value labels to each bar
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f%%', padding=3)
    
    plt.legend(title='Gender', title_fontsize=12)
    plt.tight_layout()
    plt.savefig('rq3_violations_by_gender.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # --- PART 3: Logistic Regression for Key Violation Types ---
    # Identify the most common violation type
    most_common_violation = df['Violation Type'].value_counts().index[0]
    
    print(f"\nPART 3: Logistic Regression Analysis for {most_common_violation}")
    
    # Create binary indicator for this violation type
    df['is_common_violation'] = (df['Violation Type'] == most_common_violation).astype(int)
    
    # Prepare data for logistic regression
    # Ensure categorical variables are properly encoded
    model_df = df.copy()
    
    # Create dummy variables for Race (keep all categories)
    race_dummies = pd.get_dummies(model_df['Race'], prefix='Race', drop_first=False)
    
    # Create dummy variables for Gender (keep all categories)
    gender_dummies = pd.get_dummies(model_df['Gender'], prefix='Gender', drop_first=False)
    
    # Add time of day if available
    time_features = []
    if 'Hour' in model_df.columns:
        # Create categorical time periods for easier interpretation
        model_df['TimeOfDay'] = pd.cut(
            model_df['Hour'], 
            bins=[0, 6, 12, 18, 24], 
            labels=['Night', 'Morning', 'Afternoon', 'Evening']
        )
        time_dummies = pd.get_dummies(model_df['TimeOfDay'], prefix='Time', drop_first=True)
        time_features = time_dummies.columns.tolist()
    
    # Combine features
    feature_cols = race_dummies.columns.tolist() + gender_dummies.columns.tolist() + time_features
    
    X = pd.concat([race_dummies, gender_dummies], axis=1)
    if 'Hour' in model_df.columns:
        X = pd.concat([X, time_dummies], axis=1)
    
    y = model_df['is_common_violation']
    
    # Add constant for intercept
    X = sm.add_constant(X)
    
    # Fit logistic regression
    try:
        log_reg = sm.Logit(y, X).fit(disp=0)  # disp=0 suppresses convergence messages
        
        # Extract results
        params = log_reg.params
        conf_int = log_reg.conf_int()
        odds_ratios = np.exp(params)
        odds_ratio_conf = np.exp(conf_int)
        
        # Display key results
        results_df = pd.DataFrame({
            'Odds Ratio': odds_ratios,
            'Lower 95% CI': odds_ratio_conf[0],
            'Upper 95% CI': odds_ratio_conf[1],
            'p-value': log_reg.pvalues
        })
        
        print("\nLogistic Regression Results:")
        print(f"Dependent Variable: {most_common_violation} violation (1=Yes, 0=No)")
        print(f"Number of observations: {len(model_df)}")
        print("\nOdds Ratios with 95% Confidence Intervals:")
        print(results_df.round(3))
        
        # Identify significant predictors
        sig_predictors = results_df[results_df['p-value'] < 0.05].index.tolist()
        print("\nStatistically significant predictors:")
        for pred in sig_predictors:
            if pred == 'const':
                continue
            if 'Race' in pred:
                race_group = pred.replace('Race_', '')
                effect = "more likely" if odds_ratios[pred] > 1 else "less likely"
                print(f"- {race_group} drivers are {abs(odds_ratios[pred] - 1)*100:.1f}% {effect} to have {most_common_violation} violations")
            elif 'Gender' in pred:
                gender_group = pred.replace('Gender_', '')
                effect = "more likely" if odds_ratios[pred] > 1 else "less likely"
                print(f"- {gender_group} drivers are {abs(odds_ratios[pred] - 1)*100:.1f}% {effect} to have {most_common_violation} violations")
            elif 'Time' in pred:
                time_period = pred.replace('Time_', '')
                effect = "more likely" if odds_ratios[pred] > 1 else "less likely"
                print(f"- Stops during {time_period} are {abs(odds_ratios[pred] - 1)*100:.1f}% {effect} to involve {most_common_violation} violations")
        
        # Enhanced visualization of odds ratios (excluding intercept)
        plt.figure(figsize=(12, 8))
        results_no_intercept = results_df.drop('const', errors='ignore')
        
        # Sort for better visualization and limit to top factors
        results_no_intercept = results_no_intercept.sort_values('Odds Ratio')
        
        # Define colors based on statistical significance and direction of effect
        # Red for significant negative, Blue for significant positive, Gray for non-significant
        colors = []
        for idx, p in enumerate(results_no_intercept['p-value']):
            or_value = results_no_intercept['Odds Ratio'].iloc[idx]
            if p < 0.05:
                colors.append('#E74C3C' if or_value < 1 else '#3498DB')  # Red if OR<1, Blue if OR>1
            else:
                colors.append('#95A5A6')  # Gray if not significant
        
        # Create a more appealing forest plot
        fig, ax = plt.subplots(figsize=(12, 8))
        
        # Plot points
        ax.scatter(
            results_no_intercept['Odds Ratio'],
            range(len(results_no_intercept)),
            s=100,
            color=colors,
            zorder=3
        )
        
        # Plot error bars
        for i, (idx, row) in enumerate(results_no_intercept.iterrows()):
            ax.plot(
                [row['Lower 95% CI'], row['Upper 95% CI']],
                [i, i],
                color=colors[i],
                linewidth=2,
                zorder=2
            )
            
            # Add vertical end caps
            for bound in [row['Lower 95% CI'], row['Upper 95% CI']]:
                ax.plot(
                    [bound, bound],
                    [i-0.1, i+0.1],
                    color=colors[i],
                    linewidth=2,
                    zorder=2
                )
        
        # Reference line at OR=1 (no effect)
        ax.axvline(x=1, color='black', linestyle='--', alpha=0.7, zorder=1)
        
        # Clean up axis labels
        clean_labels = []
        for label in results_no_intercept.index:
            if 'Race_' in label:
                clean_labels.append(f"Race: {label.replace('Race_', '')}")
            elif 'Gender_' in label:
                clean_labels.append(f"Gender: {label.replace('Gender_', '')}")
            elif 'Time_' in label:
                clean_labels.append(f"Time: {label.replace('Time_', '')}")
            else:
                clean_labels.append(label)
        
        # Set y-ticks
        ax.set_yticks(range(len(results_no_intercept)))
        ax.set_yticklabels(clean_labels)
        
        # Set x-axis to log scale and format labels
        ax.set_xscale('log')
        ax.xaxis.set_major_formatter(mtick.ScalarFormatter())
        
        # Add grid lines
        ax.grid(axis='x', alpha=0.3)
        
        # Add annotations for significant factors
        for i, (idx, row) in enumerate(results_no_intercept.iterrows()):
            if row['p-value'] < 0.05:
                ax.text(
                    row['Odds Ratio'] * (1.05 if row['Odds Ratio'] > 1 else 0.95),
                    i,
                    f"OR: {row['Odds Ratio']:.2f} [{row['Lower 95% CI']:.2f}-{row['Upper 95% CI']:.2f}]",
                    va='center',
                    fontsize=9,
                    fontweight='bold'
                )
        
        # Add titles and labels
        plt.title(f'Demographic Factors Associated with {most_common_violation} Violations', fontweight='bold', pad=20)
        plt.xlabel('Odds Ratio (log scale)', labelpad=10)
        plt.ylabel('Demographic Factor', labelpad=10)
        
        # Add text explanation
        fig.text(
            0.5, 0.01,
            "Odds Ratio < 1: Less likely to receive violation  |  Odds Ratio > 1: More likely to receive violation",
            ha='center', fontsize=10, style='italic'
        )
        
        # Legend for significance
        from matplotlib.lines import Line2D
        legend_elements = [
            Line2D([0], [0], marker='o', color='w', markerfacecolor='#3498DB', markersize=10, label='Significant positive association'),
            Line2D([0], [0], marker='o', color='w', markerfacecolor='#E74C3C', markersize=10, label='Significant negative association'),
            Line2D([0], [0], marker='o', color='w', markerfacecolor='#95A5A6', markersize=10, label='Non-significant association')
        ]
        ax.legend(handles=legend_elements, loc='upper right', fontsize=9)
        
        plt.tight_layout(rect=[0, 0.03, 1, 0.97])
        plt.savefig('rq3_logistic_regression_results.png', dpi=300, bbox_inches='tight')
        plt.show()
        
    except Exception as e:
        print(f"Error in logistic regression: {e}")
        print("Try fewer predictors or check for perfect separation in the data.")
    
    # --- PART 4: Heatmap visualization of Demographic and Violation Types ---
    print("\nPART 4: Heatmap Analysis of Demographic Profiles and Violations")
    
    # Create a heatmap of race and violation type relationship
    # Focus on top 10 violation types for clarity
    top10_violations = df['Violation Type'].value_counts().head(10).index.tolist()
    
    # Create a percentage-based contingency table (each row sums to 100%)
    heat_data = pd.crosstab(df['Race'], df['Violation Type'])
    heat_data_pct = heat_data.div(heat_data.sum(axis=1), axis=0) * 100
    
    # Filter to top 10 violations
    heat_data_pct = heat_data_pct[top10_violations]
    
    # Create an enhanced heatmap
    plt.figure(figsize=(14, 8))
    
    # Generate the heatmap with enhanced aesthetics
    ax = sns.heatmap(
        heat_data_pct, 
        annot=True, 
        fmt='.1f', 
        cmap='YlGnBu', 
        linewidths=0.5, 
        cbar_kws={'label': 'Percentage within Racial Group (%)'}
    )
    
    # Enhance plot appearance
    plt.title('Heatmap of Violation Types by Race (% within each racial group)', fontweight='bold', pad=20)
    plt.xlabel('Violation Type', labelpad=10)
    plt.ylabel('Race', labelpad=10)
    
    # Rotate x-axis labels for readability
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('rq3_race_violation_heatmap.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # If we have age data or vehicle age, create a third visualization
    if 'Hour' in df.columns:
        print("\nAnalysis of Violation Types by Time of Day and Demographics")
        
        # Create hour groups for better visualization
        df['Hour_Group'] = pd.cut(
            df['Hour'], 
            bins=[0, 6, 12, 18, 24], 
            labels=['Night (0-6)', 'Morning (6-12)', 'Afternoon (12-18)', 'Evening (18-24)']
        )
        
        # Create a stacked area chart of violations by hour and demographic
        # Focus on one demographic factor (race) and top 3 violations for clarity
        top3_violations = df['Violation Type'].value_counts().head(3).index.tolist()
        
        # Create the data for plotting
        plt.figure(figsize=(14, 10))
        
        # For each race, plot violation distribution by hour
        races = df['Race'].unique()
        
        # Create color palette for races
        race_colors = sns.color_palette("Set2", len(races))
        
        # Create facet grid
        fig, axes = plt.subplots(len(top3_violations), 1, figsize=(12, 4*len(top3_violations)), sharex=True)
        
        for i, violation in enumerate(top3_violations):
            # Create data frame for this violation
            violation_data = df[df['Violation Type'] == violation]
            
            # Count by hour group and race
            hour_race_counts = pd.crosstab(violation_data['Hour_Group'], violation_data['Race'])
            
            # Convert to percentages (each hour group sums to 100%)
            hour_race_pct = hour_race_counts.div(hour_race_counts.sum(axis=1), axis=0) * 100
            
            # Plot on the appropriate subplot
            hour_race_pct.plot(
                kind='bar', 
                stacked=True, 
                ax=axes[i],
                color=race_colors,
                width=0.8
            )
            
            # Customize subplot
            axes[i].set_title(f'{violation} Violations by Time of Day and Race', fontweight='bold')
            axes[i].set_ylabel('Percentage within Time Period (%)')
            axes[i].grid(axis='y', alpha=0.3)
            
            # Only show legend on the last subplot
            if i == len(top3_violations) - 1:
                axes[i].legend(title='Race', bbox_to_anchor=(1.05, 1), loc='upper left')
            else:
                axes[i].get_legend().remove()
        
        # Set common x-axis label
        fig.text(0.5, 0.04, 'Time of Day', ha='center', va='center', fontsize=12)
        
        plt.tight_layout()
        plt.savefig('rq3_violation_by_time_race.png', dpi=300, bbox_inches='tight')
        plt.show()

print("\nResearch Question 3 Analysis Complete")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import matplotlib.ticker as mtick
import matplotlib.dates as mdates
from matplotlib.colors import LinearSegmentedColormap
import calendar

# Set plot style for better aesthetics
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Start with the cleaned dataset
df = df_cleaned.copy()

print("RQ4: What is the relationship between time, date and type of traffic violations issued?")
print("-" * 80)

# Check for required columns
required_columns = ['Hour', 'Day_of_Week', 'Month', 'Year_of_Stop', 'Violation Type']
missing_cols = [col for col in required_columns if col not in df.columns]

if missing_cols:
    print(f"Warning: Missing time/date columns: {missing_cols}")
    print("Analysis will be limited to available time/date information.")

# --- PART 1: Hour of Day Analysis ---
if 'Hour' in df.columns:
    print("\nPART 1: Analysis of Traffic Violations by Hour of Day")
    
    # Create contingency table
    hour_violation = pd.crosstab(df['Hour'], df['Violation Type'])
    
    # Chi-square test
    chi2, p, dof, expected = chi2_contingency(hour_violation)
    print(f"\nChi-square test for Hour vs. Violation Type:")
    print(f"Chi-square statistic: {chi2:.2f}")
    print(f"p-value: {p:.8f}")
    print(f"Degrees of freedom: {dof}")
    print(f"Conclusion: {'Significant association' if p < 0.05 else 'No significant association'} between hour of day and violation types")
    
    # Get total violations by hour
    hourly_counts = df.groupby('Hour').size()
    
    # Get top 5 violation types
    top_violations = df['Violation Type'].value_counts().head(5).index.tolist()
    
    # Create hourly patterns visualization
    plt.figure(figsize=(14, 8))
    
    # Plot total violations by hour
    ax1 = plt.gca()
    ax1.bar(hourly_counts.index, hourly_counts.values, alpha=0.3, color='gray', label='All Violations')
    ax1.set_xlabel('Hour of Day (24-hour format)', labelpad=10)
    ax1.set_ylabel('Number of Violations', labelpad=10)
    ax1.set_xticks(range(0, 24))
    ax1.grid(True, linestyle='--', alpha=0.3)
    
    # Create second y-axis for percentages
    ax2 = ax1.twinx()
    
    # Plot each top violation type as percentage of hourly total
    for violation in top_violations:
        if violation in hour_violation.columns:
            # Calculate percentage of this violation in each hour
            hourly_pct = (hour_violation[violation] / hour_violation.sum(axis=1)) * 100
            ax2.plot(hourly_pct.index, hourly_pct.values, marker='o', linewidth=2, label=violation)
    
    ax2.set_ylabel('Percentage of Hourly Violations (%)', labelpad=10)
    ax2.yaxis.set_major_formatter(mtick.PercentFormatter())
    
    # Combine legends
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.15), 
              ncol=3, frameon=True, title='Violation Types')
    
    plt.title('Traffic Violations by Hour of Day', fontweight='bold', pad=20)
    plt.tight_layout()
    plt.savefig('rq4_violations_by_hour.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Identify peak hours for different violation types
    print("\nPeak hours for top violation types:")
    for violation in top_violations:
        if violation in hour_violation.columns:
            # Calculate percentage of this violation in each hour
            hourly_pct = (hour_violation[violation] / hour_violation.sum(axis=1)) * 100
            peak_hour = hourly_pct.idxmax()
            print(f"  {violation}: Peak at {peak_hour}:00 ({hourly_pct[peak_hour]:.1f}% of violations in that hour)")

# --- PART 2: Day of Week Analysis ---
if 'Day_of_Week' in df.columns:
    print("\nPART 2: Analysis of Traffic Violations by Day of Week")
    
    # Ensure proper day ordering
    if df['Day_of_Week'].dtype == object:
        # If days are strings, order them properly
        day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        
        # Check if our dataset uses these exact labels
        if set(df['Day_of_Week'].unique()).issubset(set(day_order)):
            df['Day_of_Week'] = pd.Categorical(df['Day_of_Week'], categories=day_order, ordered=True)
    else:
        # If days are numeric (0-6), add labels for readability
        if df['Day_of_Week'].max() <= 6:
            day_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 
                       3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
            df['Day_Name'] = df['Day_of_Week'].map(day_map)
            day_col = 'Day_Name'
        else:
            day_col = 'Day_of_Week'
    
    # Use the appropriate day column
    day_col = 'Day_of_Week' if 'Day_Name' not in df.columns else 'Day_Name'
    
    # Create contingency table
    day_violation = pd.crosstab(df[day_col], df['Violation Type'])
    
    # Chi-square test
    chi2, p, dof, expected = chi2_contingency(day_violation)
    print(f"\nChi-square test for Day of Week vs. Violation Type:")
    print(f"Chi-square statistic: {chi2:.2f}")
    print(f"p-value: {p:.8f}")
    print(f"Degrees of freedom: {dof}")
    print(f"Conclusion: {'Significant association' if p < 0.05 else 'No significant association'} between day of week and violation types")
    
    # Calculate percentages for visualization
    day_violation_pct = pd.DataFrame()
    
    for violation in top_violations:
        if violation in day_violation.columns:
            # Calculate percentage within each day
            day_violation_pct[violation] = (day_violation[violation] / day_violation.sum(axis=1)) * 100
    
    # Visualization of day of week patterns
    plt.figure(figsize=(12, 7))
    
    # Create heat map of day of week vs violation type
    ax = sns.heatmap(day_violation_pct, cmap='YlGnBu', annot=True, fmt='.1f', 
                    linewidths=0.5, cbar_kws={'label': 'Percentage of Daily Violations (%)'})
    
    plt.title('Distribution of Violation Types by Day of Week', fontweight='bold', pad=20)
    plt.xlabel('Violation Type', labelpad=10)
    plt.ylabel('Day of Week', labelpad=10)
    plt.tight_layout()
    plt.savefig('rq4_violations_by_day.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Compare weekday vs weekend patterns
    if df[day_col].dtype == object:
        df['is_weekend'] = df[day_col].isin(['Saturday', 'Sunday'])
    else:
        df['is_weekend'] = df[day_col].isin([5, 6])  # Assuming 0=Monday, 6=Sunday
    
    weekday_weekend = pd.crosstab(df['is_weekend'], df['Violation Type'])
    weekday_weekend_pct = (weekday_weekend.div(weekday_weekend.sum(axis=1), axis=0)) * 100
    
    print("\nWeekday vs. Weekend Comparison:")
    for violation in top_violations:
        if violation in weekday_weekend_pct.columns:
            weekday_pct = weekday_weekend_pct.loc[False, violation] if False in weekday_weekend_pct.index else 0
            weekend_pct = weekday_weekend_pct.loc[True, violation] if True in weekday_weekend_pct.index else 0
            diff = weekend_pct - weekday_pct
            print(f"  {violation}: {weekday_pct:.1f}% on weekdays, {weekend_pct:.1f}% on weekends ({diff:+.1f}% difference)")

# --- PART 3: Month and Seasonal Analysis ---
if 'Month' in df.columns:
    print("\nPART 3: Analysis of Traffic Violations by Month")
    
    # Ensure proper month ordering
    if df['Month'].dtype == object:
        # If months are strings, order them properly
        month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
                      'July', 'August', 'September', 'October', 'November', 'December']
        
        # Check if our dataset uses these exact labels
        if set(df['Month'].unique()).issubset(set(month_order)):
            df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
    else:
        # If months are numeric (1-12), add labels for readability
        if df['Month'].max() <= 12:
            month_map = {i: calendar.month_name[i] for i in range(1, 13)}
            df['Month_Name'] = df['Month'].map(month_map)
            month_col = 'Month_Name'
        else:
            month_col = 'Month'
    
    # Use the appropriate month column
    month_col = 'Month' if 'Month_Name' not in df.columns else 'Month_Name'
    
    # Create contingency table
    month_violation = pd.crosstab(df[month_col], df['Violation Type'])
    
    # Chi-square test
    chi2, p, dof, expected = chi2_contingency(month_violation)
    print(f"\nChi-square test for Month vs. Violation Type:")
    print(f"Chi-square statistic: {chi2:.2f}")
    print(f"p-value: {p:.8f}")
    print(f"Degrees of freedom: {dof}")
    print(f"Conclusion: {'Significant association' if p < 0.05 else 'No significant association'} between month and violation types")
    
    # Add season information
    if df['Month'].dtype == object:
        winter = ['December', 'January', 'February']
        spring = ['March', 'April', 'May']
        summer = ['June', 'July', 'August']
        fall = ['September', 'October', 'November']
        
        df['Season'] = 'Unknown'
        df.loc[df['Month'].isin(winter), 'Season'] = 'Winter'
        df.loc[df['Month'].isin(spring), 'Season'] = 'Spring'
        df.loc[df['Month'].isin(summer), 'Season'] = 'Summer'
        df.loc[df['Month'].isin(fall), 'Season'] = 'Fall'
    else:
        df['Season'] = 'Unknown'
        df.loc[df['Month'].isin([12, 1, 2]), 'Season'] = 'Winter'
        df.loc[df['Month'].isin([3, 4, 5]), 'Season'] = 'Spring'
        df.loc[df['Month'].isin([6, 7, 8]), 'Season'] = 'Summer'
        df.loc[df['Month'].isin([9, 10, 11]), 'Season'] = 'Fall'
    
    # Seasonal analysis
    season_violation = pd.crosstab(df['Season'], df['Violation Type'])
    season_violation_pct = (season_violation.div(season_violation.sum(axis=1), axis=0)) * 100
    
    # Create seasonal visualization
    plt.figure(figsize=(14, 8))
    
    # For each top violation, plot its seasonal pattern
    bar_width = 0.15
    r = np.arange(len(season_violation_pct.columns[:10]))  # Limit to top 10 violations
    
    # Set season colors
    season_colors = {'Winter': '#6495ED', 'Spring': '#98FB98', 
                    'Summer': '#FFA07A', 'Fall': '#DAA520'}
    
    # Plot bars for each season
    for i, season in enumerate(season_violation_pct.index):
        if season in season_colors:
            plt.bar(r + i*bar_width, 
                   season_violation_pct.iloc[i, :10], 
                   width=bar_width, 
                   label=season,
                   color=season_colors[season],
                   alpha=0.8)
    
    # Add labels
    plt.xlabel('Violation Type', labelpad=10)
    plt.ylabel('Percentage within Season (%)', labelpad=10)
    plt.title('Seasonal Patterns in Traffic Violations', fontweight='bold', pad=20)
    plt.xticks(r + bar_width*1.5, season_violation_pct.columns[:10], rotation=45, ha='right')
    plt.legend(title='Season')
    plt.grid(axis='y', alpha=0.3)
    
    # Format y-axis as percentage
    plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
    
    plt.tight_layout()
    plt.savefig('rq4_violations_by_season.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Identify seasonal patterns for top violations
    print("\nSeasonal patterns for top violation types:")
    for violation in top_violations:
        if violation in season_violation_pct.columns:
            peak_season = season_violation_pct[violation].idxmax()
            low_season = season_violation_pct[violation].idxmin()
            print(f"  {violation}: Highest in {peak_season} ({season_violation_pct[violation][peak_season]:.1f}%), " + 
                 f"lowest in {low_season} ({season_violation_pct[violation][low_season]:.1f}%)")

# --- PART 4: Combined Hour and Day Analysis (Heatmap) ---
if all(col in df.columns for col in ['Hour', 'Day_of_Week']):
    print("\nPART 4: Combined Analysis of Hour and Day of Week")
    
    # Create day column for plotting
    day_col = 'Day_of_Week' if 'Day_Name' not in df.columns else 'Day_Name'
    
    # For the most common violation type, create an hour-day heatmap
    most_common_violation = df['Violation Type'].value_counts().index[0]
    
    # Filter data for this violation
    violation_df = df[df['Violation Type'] == most_common_violation]
    
    # Create pivot table for heatmap
    hour_day_pivot = pd.crosstab(violation_df['Hour'], violation_df[day_col])
    
    # Normalize to show percentage of total violations
    hour_day_pct = (hour_day_pivot / hour_day_pivot.values.sum()) * 100
    
    # Create custom colormap that works well for percentages
    colors = ['#f7fbff', '#deebf7', '#c6dbef', '#9ecae1', '#6baed6', '#4292c6', '#2171b5', '#08519c', '#08306b']
    cmap = LinearSegmentedColormap.from_list('blue_gradient', colors)
    
    plt.figure(figsize=(14, 10))
    
    # Plot heatmap
    ax = sns.heatmap(hour_day_pct, cmap=cmap, annot=True, fmt='.2f', linewidths=0.5,
                    cbar_kws={'label': f'Percentage of Total {most_common_violation} Violations (%)'})
    
    plt.title(f'Distribution of {most_common_violation} Violations by Hour and Day', fontweight='bold', pad=20)
    plt.xlabel('Day of Week', labelpad=10)
    plt.ylabel('Hour of Day (24-hour format)', labelpad=10)
    
    # Ensure proper ordering of days if categorical
    if df[day_col].dtype.name == 'category':
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    else:
        plt.xticks(rotation=45, ha='right')
    
    plt.tight_layout()
    plt.savefig('rq4_hour_day_heatmap.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Identify highest-risk time slots
    flat_idx = hour_day_pct.values.flatten().argsort()[-5:][::-1]  # Get indices of top 5 values
    rows, cols = np.unravel_index(flat_idx, hour_day_pct.shape)
    
    print(f"\nPeak time slots for {most_common_violation} violations:")
    for i in range(len(rows)):
        hour = hour_day_pct.index[rows[i]]
        day = hour_day_pct.columns[cols[i]]
        pct = hour_day_pct.iloc[rows[i], cols[i]]
        print(f"  {day} at {hour}:00 - {pct:.2f}% of all {most_common_violation} violations")

# --- PART 5: Year-over-Year Trends (if multiple years of data) ---
if 'Year_of_Stop' in df.columns and df['Year_of_Stop'].nunique() > 1:
    print("\nPART 5: Year-over-Year Trends in Traffic Violations")
    
    # Create yearly trend analysis
    year_counts = df.groupby(['Year_of_Stop', 'Violation Type']).size().unstack(fill_value=0)
    
    # Calculate percentage within each year
    year_pct = year_counts.div(year_counts.sum(axis=1), axis=0) * 100
    
    # Track trends for top violations
    print("\nYear-over-year trends for top violation types:")
    for violation in top_violations:
        if violation in year_pct.columns:
            trend_data = year_pct[violation]
            years = list(trend_data.index)
            if len(years) > 1:
                first_year = years[0]
                last_year = years[-1]
                first_pct = trend_data[first_year]
                last_pct = trend_data[last_year]
                change = last_pct - first_pct
                print(f"  {violation}: {first_pct:.1f}% in {first_year} → {last_pct:.1f}% in {last_year} ({change:+.1f}% change)")
    
    # Visualization of yearly trends
    plt.figure(figsize=(14, 8))
    
    # Plot trends for top violations
    for violation in top_violations:
        if violation in year_pct.columns:
            plt.plot(year_pct.index, year_pct[violation], marker='o', linewidth=2, markersize=8, label=violation)
    
    plt.title('Trends in Traffic Violation Types Over Time', fontweight='bold', pad=20)
    plt.xlabel('Year', labelpad=10)
    plt.ylabel('Percentage of Annual Violations (%)', labelpad=10)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.legend(title='Violation Type')
    
    # Format y-axis as percentage
    plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
    
    # Format x-axis as years
    plt.xticks(year_pct.index)
    
    plt.tight_layout()
    plt.savefig('rq4_violation_trends.png', dpi=300, bbox_inches='tight')
    plt.show()

print("\nResearch Question 4 Analysis Complete")
```
